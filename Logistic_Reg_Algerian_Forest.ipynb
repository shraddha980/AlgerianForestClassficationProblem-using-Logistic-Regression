## Logistic Regression


Logistic Regression is a supervised classification technique used to classify labelled data. Classification can be done on
Binary Labelled Output. Logistic Regression is primary Binary Classification Technique.

Logistic Regression applies Sigmoid function on the regression line to get the binary output by using a threshold. It uses the cost function and repeat convergence algorithm to get the values of coefficients and forms a gradient descent.

We can determine the performance of logistic Regression using Accuracy Rate, Precision, Recall, Sensitivity, RUC Curve 
and confusion matrix

## Implementation of Logistic Regression

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

Loading Algerian Forest Dataset

```python
df= pd.read_csv("Algerian_forest_fires_dataset.csv")
```

Checking initial 10 rows of dataset

```python
df.head(10)
```

Checking the no of rows and columns in the data

```python
df.shape
```

Checking the statistics for the dataset

df.describe()

Checing the Correlation strength between predictors

```python
df.corr()
```

Separating Predictor and Target Variables

```python
x=df.iloc[:,0:13]
```

```python
y=df.iloc[:,-1]
```

```python
x
```

```python
y= y.str.replace(" ", "")
```

```python
y
```

Splitting x and y into train and test variables

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size = 0.25, random_state = 0)
```

Transforming variables by importing standard scaler applying it on train and test variable

```python
sc= StandardScaler()
xtrain = sc.fit_transform(X_train) 
xtest = sc.transform(X_test)
```

Importing Logistic Regression and applying it on xtrain and ytrain

```python
from sklearn.linear_model import LogisticRegression
```

```python
classifier = LogisticRegression(random_state=0)
classifier.fit(xtrain,y_train)
```

Predicting the output

```python
y_pred = classifier.predict(xtest)
```

Checking the Performance of our model

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test, y_pred)
```

```python
confusion_matrix(y_test, y_pred)
```
